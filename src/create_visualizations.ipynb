# Create Visualizations

This script takes the data in 'processed_data' and turns them into line plots.

## Import Packages

```python
import pandas as pd
import seaborn as sns
```

## Load Data

```python
mean_df = pd.read_csv('../processed_data/average_viewcount.csv')
peak_df = pd.read_csv('../processed_data/peak_viewcount.csv')
newest_df = pd.read_csv('../processed_data/fewest_months_viewcount.csv')
```

## Create Visualizations

### Visualization 1

```python
sns.set_style("white")
plot1 = sns.lineplot(data=mean_df, x='month', y='views', hue='article')
plot1.set(title='Wikipedia Dinosaur Article Viewcounts\nArticles with Lowest or Highest Average Viewcounts')
plot1.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
for ind, label in enumerate(plot1.get_xticklabels()):
    if ind % 10 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
```

### Visualization 2

```python
sns.set_style("white")
plot2 = sns.lineplot(data=peak_df, x='month', y='views', hue='article', style='access_type')
plot2.set(title='Wikipedia Dinosaur Article Viewcounts\nArticles with Highest Single Month Viewcounts')
plot2.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
for ind, label in enumerate(plot2.get_xticklabels()):
    if ind % 10 == 0:  # every 10th label is kept
        label.set_visible(True)
    else:
        label.set_visible(False)
```

### Visualization 3

```python
sns.set_style("white")
plot3 = sns.lineplot(data=newest_df, x='month', y='views', hue='article', style='access_type')
plot3.set(title='Wikipedia Dinosaur Article Viewcounts\nArticles with Fewest Months of Data')
plot3.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
```

## Save Visualizations

```python
plot1.get_figure().savefig('../result_images/average_viewcount.png', bbox_inches='tight')
plot2.get_figure().savefig('../result_images/peak_viewcount.png', bbox_inches='tight')
plot3.get_figure().savefig('../result_images/fewest_months_viewcount.png', bbox_inches='tight')
```
